## 1. Descripción

BIRCH, que significa "Balanced Iterative Reducing and Clustering Using Hierarchies" (Reducción e Agrupamiento Equilibrado Iterativo Usando Jerarquías), es un algoritmo de agrupamiento (clustering) que se utiliza para dividir un conjunto de datos en grupos o clústeres basándose en características similares. Fue propuesto por Tian Zhang, Raghu Ramakrishnan y Miron Livny en 1996.

Pasos del Algoritmo:

1. **Construcción del Árbol:** BIRCH comienza con un árbol vacío y agrega puntos de datos uno por uno. A medida que se agregan los puntos, se actualiza la estructura del árbol, dividiendo los nodos cuando sea necesario.

2.** Agrupamiento:** Cuando se ha construido el árbol, se pueden extraer los clústeres de diferentes niveles de granularidad según las necesidades del usuario. Los puntos de datos se agrupan en clústeres basados en su ubicación en el árbol.

## 5. Supuestos y Restricciones:

Supuestos:

1. Distribución equilibrada de clústeres: BIRCH asume que los clústeres tienen una distribución equilibrada en términos de tamaño. En otras palabras, se espera que los clústeres tengan tamaños similares.

2. Tamaño máximo del clúster: El algoritmo requiere que se especifique un tamaño máximo del clúster, denominado "LF" (Factor de Carga Límite). Este parámetro controla cuántos puntos de datos pueden caber en un clúster antes de que se divida en subclústeres más pequeños. El valor de LF es importante y debe ajustarse cuidadosamente para que BIRCH funcione correctamente.

3. Similitud de los datos: BIRCH se basa en la métrica de distancia para medir la similitud entre puntos de datos. Por lo tanto, asume que la métrica de distancia utilizada es apropiada para el tipo de datos y el problema.

4. Estructura jerárquica: El algoritmo asume que los datos pueden organizarse en una estructura jerárquica de árbol, lo que significa que se pueden formar clústeres a diferentes niveles de granularidad.

4. Compresión de datos: BIRCH utiliza técnicas de compresión de datos para representar de manera compacta la información de cada clúster en forma de "entradas de clúster". Se supone que estas representaciones resumen adecuadamente los datos en cada nodo del árbol.

5. Número de clústeres desconocido: BIRCH no requiere que se especifique previamente el número de clústeres (K) en los datos. En cambio, el árbol BIRCH se construye de manera adaptativa a medida que se procesan los datos.

Restricciones:

1. Requiere una cantidad significativa de memoria: A pesar de su capacidad para manejar conjuntos de datos grandes en un enfoque offline, BIRCH todavía necesita suficiente memoria para mantener la estructura del árbol BIRCH y las entradas de clúster en memoria. En conjuntos de datos extremadamente grandes, el uso de memoria puede ser una restricción.

2. Sensible a la elección de hiperparámetros: BIRCH requiere la elección de hiperparámetros como el "Factor de Carga Límite" (LF) y el número de dimensiones para la métrica de distancia. La elección inadecuada de estos hiperparámetros puede afectar significativamente los resultados del clustering.

3. Requiere una métrica de distancia adecuada: BIRCH utiliza una métrica de distancia para medir la similitud entre puntos de datos. La elección de la métrica de distancia debe ser apropiada para el tipo de datos y el problema que estás abordando.

4. No es adecuado para clústeres con formas no esféricas: BIRCH asume que los clústeres tienen una forma esférica o aproximadamente esférica. Si los clústeres tienen formas muy irregulares o no esféricas, BIRCH puede tener dificultades para encontrar clústeres precisos.

5. No proporciona el número óptimo de clústeres: Aunque BIRCH no requiere especificar el número de clústeres de antemano, no proporciona automáticamente el número óptimo de clústeres. La interpretación de los resultados y la selección del número de clústeres pueden requerir conocimiento adicional del dominio.

6. Sensible a los valores atípicos (outliers): BIRCH puede ser sensible a valores atípicos o datos ruidosos en el conjunto de datos, lo que puede afectar negativamente la calidad de los clústeres.

7. Puede no ser eficiente en alta dimensionalidad: En conjuntos de datos con muchas dimensiones (alta dimensionalidad), la eficiencia de BIRCH puede verse comprometida debido a la maldición de la dimensionalidad.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('wine-clustering.csv')
df.sample(2)
```

## Anális de DATOS:

```python
# Resumen estadístico
print(df.describe())

# Histogramas para visualizar las distribuciones
df.hist(bins=10, figsize=(15, 10))
plt.show()

# Matriz de correlación
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
plt.matshow(correlation_matrix, cmap='coolwarm', fignum=1)
plt.colorbar()
plt.xticks(range(len(df.columns)), df.columns, rotation='vertical')
plt.yticks(range(len(df.columns)), df.columns)
plt.show()
```

## Aplicando Algoritmo

```python

from sklearn.cluster import Birch
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


# Dividir los datos en conjuntos de entrenamiento y prueba (20% para prueba)
X_train, X_test = train_test_split(df, test_size=0.2, random_state=0)

# Crear un objeto BIRCH con los parámetros deseados
birch = Birch(n_clusters=3, threshold=0.5)

# Entrenar el modelo en el conjunto de entrenamiento
birch.fit(X_train)

# Obtener las etiquetas de clúster para el conjunto de prueba
labels = birch.predict(X_test)

# Graficar los resultados (por ejemplo, si tus datos tienen solo dos características)
plt.scatter(X_test.iloc[:, 0], X_test.iloc[:, 1], c=labels, cmap='rainbow')
plt.title("Resultados del Clustering BIRCH")
plt.xlabel("x 1")
plt.ylabel("y 2")
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Calcular el índice de silueta
silhouette_avg = silhouette_score(X_test, labels)
```

```python
# Calcular el índice Davies-Bouldin
davies_bouldin = davies_bouldin_score(X_test, labels)

# Calcular el índice Calinski-Harabasz
calinski_harabasz = calinski_harabasz_score(X_test, labels)

print("Índice de Silueta:", silhouette_avg)
print("Índice Davies-Bouldin:", davies_bouldin)
print("Índice Calinski-Harabasz:", calinski_harabasz)




```
