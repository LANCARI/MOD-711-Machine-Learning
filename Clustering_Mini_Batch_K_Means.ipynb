## 1. Descripción

El algoritmo Mini-Batch K-Means es una variante del algoritmo K-Means que aborda la eficiencia computacional al acelerar el proceso de convergencia utilizando muestras aleatorias (mini-batches) de datos en lugar de utilizar el conjunto completo de datos en cada iteración. A continuación, se describe el algoritmo de Mini-Batch K-Means:

1. Inicialización de Centroides:

Selecciona K centroides iniciales de manera aleatoria del conjunto de datos o utilizando algún otro método de inicialización. Estos centroides representan el prototipo de cada clúster.
2. Mini-Batch Aleatorio:

En cada iteración, el algoritmo selecciona aleatoriamente un mini-batch (una muestra de tamaño fijo) de los datos de entrada en lugar de utilizar todos los datos en cada paso. Esta es la principal diferencia entre Mini-Batch K-Means y K-Means convencional.
3. Asignación de Puntos al Clúster más Cercano:

Para cada punto en el mini-batch, calcula la distancia a los centroides actuales y asigna el punto al clúster cuyo centroide está más cerca.
4. Actualización de Centroides:

Después de asignar los puntos al clúster más cercano en el mini-batch, actualiza los centroides de cada clúster calculando el promedio de los puntos asignados a ese clúster en el mini-batch. Esto se hace para todos los clústeres.
5. Actualización de Todos los Datos:

Ocasionalmente (después de un cierto número de mini-batches), los centroides se actualizan utilizando todos los datos disponibles en lugar de solo el mini-batch actual. Esto ayuda a estabilizar la convergencia y mejorar la precisión.
6. Convergencia:

El algoritmo continúa iterando a través de mini-batches y actualizando los centroides hasta que se cumple un criterio de convergencia, como un número máximo de iteraciones o una pequeña variación en los centroides.
7. Resultados:

Una vez que el algoritmo ha convergido, los centroides representan los clústeres finales, y los puntos se asignan a los clústeres en función de la distancia a los centroides.
El algoritmo Mini-Batch K-Means se utiliza para acelerar la convergencia en conjuntos de datos grandes y suele ser más rápido que el K-Means convencional sin sacrificar significativamente la calidad de los clústeres. Sin embargo, debido al uso de mini-batches aleatorios, la calidad de la solución puede ser ligeramente inferior a la de K-Means en conjuntos de datos más pequeños. La elección del tamaño del mini-batch es un hiperparámetro importante que puede afectar la convergencia y la calidad de los resultados.

## 5. Supuestos y Restricciones:

Supuestos:

1. Número de Clústeres (K): Debes especificar el número de clústeres (K) antes de ejecutar el algoritmo. Esto implica que se asume que el número de clústeres es conocido de antemano o ha sido determinado mediante algún criterio de selección adecuado.

2. Inicialización de Centroides: El algoritmo parte del supuesto de que se proporciona una inicialización adecuada para los centroides. La elección de los centroides iniciales puede influir en la convergencia y la calidad de la solución final.

3. Homogeneidad Dentro de Clústeres: Se supone que los puntos dentro de un mismo clúster son homogéneos en términos de similitud o distancia entre ellos. El objetivo es minimizar la varianza intraclúster, lo que significa que se espera que los puntos dentro del mismo clúster sean más similares entre sí que con los puntos de otros clústeres.

4. Varianza Equivalente: El algoritmo K-Means asume que la varianza dentro de cada clúster es aproximadamente la misma. Esto significa que se supone que los clústeres tienen una forma esférica y tamaños relativamente iguales.

5. Independencia de Atributos: El algoritmo asume que los atributos o características utilizados para medir la similitud entre puntos son independientes. Esto puede ser problemático en conjuntos de datos donde las características no son independientes en la realidad.

6. Convergencia: Se asume que el algoritmo convergerá a una solución óptima o cercana a la óptima después de un número finito de iteraciones. Sin embargo, no hay garantía de que siempre converja a la solución globalmente óptima.

7. Criterio de Parada: El algoritmo requiere un criterio de parada, como un número máximo de iteraciones o una pequeña variación en los centroides, para determinar cuándo ha convergido. La elección de este criterio puede afectar la convergencia y los resultados.

8. Tamaño del Mini-Batch: En el caso específico del Mini-Batch K-Means, se supone que el tamaño del mini-batch utilizado es adecuado y afectará la velocidad de convergencia y la calidad de la solución.


Restricciones:

1. Dependencia del Tamaño del Mini-Batch: El rendimiento y la calidad de la solución del algoritmo Mini-Batch K-Means pueden depender del tamaño del mini-batch utilizado. Un tamaño de mini-batch muy pequeño puede conducir a una convergencia más rápida pero a una solución menos precisa, mientras que un tamaño de mini-batch demasiado grande puede ralentizar la convergencia y requerir más memoria.

2. Dependencia de la Inicialización: La elección de la inicialización de los centroides iniciales sigue siendo un aspecto crítico del algoritmo. Una mala elección inicial puede afectar la calidad de la solución final y la velocidad de convergencia.

3. Número de Clústeres (K): Al igual que en el K-Means tradicional, debes especificar el número de clústeres (K) antes de ejecutar el algoritmo. Elegir el valor correcto de K puede ser un desafío y puede requerir métodos de validación de clústeres.

4. Convergencia no Garantizada: Aunque el algoritmo Mini-Batch K-Means generalmente converge a una solución cercana a la óptima, no hay garantía de convergencia a la solución globalmente óptima. La calidad de la solución puede depender de la inicialización y el número de iteraciones.

5. Sensibilidad al Tamaño de Datos: El algoritmo Mini-Batch K-Means es especialmente útil para conjuntos de datos grandes, pero puede ser menos efectivo en conjuntos de datos pequeños debido a la variabilidad introducida por el muestreo aleatorio de mini-batches.

6. Requerimiento de Parámetros: Además del número de clústeres (K), el algoritmo tiene otros parámetros como el tamaño del mini-batch, que deben ser ajustados adecuadamente. En la práctica, encontrar los valores óptimos de estos parámetros puede requerir experimentación.

7. Ruido en Mini-Batches: Dado que se utiliza una muestra aleatoria de datos en cada iteración, los mini-batches pueden contener ruido que afecta la convergencia y la calidad de la solución. Esto puede ser especialmente relevante en conjuntos de datos con datos atípicos o valores atípicos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('wine-clustering.csv')
df.sample(2)
```

## Anális de DATOS:

```python
# Resumen estadístico
print(df.describe())

# Histogramas para visualizar las distribuciones
df.hist(bins=10, figsize=(15, 10))
plt.show()

# Matriz de correlación
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
plt.matshow(correlation_matrix, cmap='coolwarm', fignum=1)
plt.colorbar()
plt.xticks(range(len(df.columns)), df.columns, rotation='vertical')
plt.yticks(range(len(df.columns)), df.columns)
plt.show()
```

## Aplicando Algoritmo

```python
# Importa las bibliotecas necesarias
import numpy as np
from sklearn.cluster import MiniBatchKMeans
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt



# Divide los datos en conjuntos de entrenamiento y prueba (70% entrenamiento, 30% prueba)
X_train, X_test = train_test_split(df, test_size=0.3, random_state=0)

# Especifica el número de clústeres (K)
n_clusters = 3

# Inicializa el modelo Mini-Batch K-Means
model = MiniBatchKMeans(n_clusters=n_clusters, random_state=0)

# Entrena el modelo en los datos de entrenamiento
model.fit(X_train)

# Obtiene las etiquetas de clústeres para cada punto en los datos de prueba
labels = model.predict(X_test)

# Visualiza los resultados (por ejemplo, en función de dos atributos)
atributo1 = 'Alcohol'
atributo2 = 'Malic_Acid'

plt.figure(figsize=(10, 6))
for cluster_label in range(n_clusters):
    cluster_points = X_test[labels == cluster_label]
    plt.scatter(
        cluster_points[atributo1],
        cluster_points[atributo2],
        label=f'Cluster {cluster_label}',
    )

plt.xlabel(atributo1)
plt.ylabel(atributo2)
plt.title('Resultados del Mini-Batch K-Means')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# Calcular el índice de silueta para evaluar la calidad de los clústeres
silhouette_avg = silhouette_score(X_test, labels)
print("Índice de Silueta:", silhouette_avg)

# Calcular el coeficiente de Calinski-Harabasz para evaluar la calidad de los clústeres
calinski_harabasz = calinski_harabasz_score(X_test, labels)
print("Coeficiente de Calinski-Harabasz:", calinski_harabasz)

# Calcular el coeficiente de Davies-Bouldin para evaluar la calidad de los clústeres
davies_bouldin = davies_bouldin_score(X_test, labels)
print("Coeficiente de Davies-Bouldin:", davies_bouldin)
```
